```python
complaints = pd.read_excel('complaints.xlsx')
```

```python
complaints.head()
```

complaints.groupby('facility_id')

```python
com_lite = complaints[['facility_id', 'complaint_id']]
```

```python
com_per_fac = com_lite.groupby('facility_id').count().reset_index()
```

```python
com_per_fac = com_per_fac.rename(columns = {'complaint_id': 'complaint_count'})
```

```python
com_per_fac.sort_values('complaint_count', ascending=False)
```

```python
df.merge(com_per_fac, left_on='facid', right_on='facility_id', how='left')
```

```python
fac_join = df.merge(com_per_fac, left_on='facid', right_on='facility_id', how='left')
```

```python
fac_join.head()
```

```python
fac_join.complaint_count/fac_join.capacity
```

```python
fac_join['comp_rate'] = fac_join.complaint_count/fac_join.capacity
```

```python
fac_join.head()
```

```python
fac_join.sort_values('comp_rate', ascending=False)
```

```python
fac_join['comp_rate'].mean()
```
